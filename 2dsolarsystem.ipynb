```python
import matplotlib.pyplot as plt #needed for plotting
import numpy as np #needed for arrays
import cmath #import cmath library to use complex numbers
import math #needed for pi
import matplotlib #needed for plotting

k=10000*60
z=4*((math.pi)**2)
MJS=z*(1.9e27/2e30)
MMS=z*(3.3e23/2e30)
MES=z*(6e24/2e30)
MVS=z*(4.9e24/2e30)
MMarS=z*(6.4e23/2e30)
MSS=z*(5.7e26/2e30)
MUS=z*(8.7e25/2e30)
MNS=z*(1e26/2e30)



dt=0.0001
pi2=(math.pi)**2
#initial conditions for earth at midnight March 22 2020 taken from jpl website
vxe=np.zeros(k, dtype=float)+(1.835374837557676E-04*365) 

vye=np.zeros(k, dtype=float)+(-1.726909116124397E-02*365) 


xe=np.zeros(k, dtype=float)+-9.960757640593636E-01 

ye=np.zeros(k, dtype=float)+-2.704940063009914E-02 


#intial conditions for jupiter at midnight March 22 2020 taken from jpl website

vxj=np.zeros(k, dtype=float)+(7.283728058285744E-03*365)

vyj=np.zeros(k, dtype=float)+(1.987713648940754E-03*365)

xj=np.zeros(k, dtype=float)+1.122552745194473

yj=np.zeros(k, dtype=float)+-5.075240594315544

#initial conditions for mercury at midnight March 22 2020 taken from jpl website

xmer=np.zeros(k, dtype=float)+-2.101621875192881E-01

ymer=np.zeros(k, dtype=float)+-4.122357016287996E-01

vxmer=np.zeros(k, dtype=float)+(1.939765286815084E-02*365)

vymer=np.zeros(k, dtype=float)+(-1.141656181025629E-02*365)

#initial conditions for venus at midnight March 22 2020 taken from jpl website
xv=np.zeros(k, dtype=float)+-5.030556915097480E-01

yv=np.zeros(k, dtype=float)+5.116773131958444E-01

vxv=np.zeros(k, dtype=float)+(-1.449598844734853E-02*365)

vyv=np.zeros(k, dtype=float)+(-1.428649516902965E-02*365)

#initial conditions for mars at midnight March 22 2020 taken from jpl website

xmar=np.zeros(k, dtype=float)+-3.792446972622535E-01

ymar=np.zeros(k, dtype=float)+-1.439103434928850

vxmar=np.zeros(k, dtype=float)+(1.406015061900062E-02*365)

vymar=np.zeros(k, dtype=float)+(-2.363372111999742E-03*365)

#initial conditions for saturn at midnight March 22 2020 taken from jpl website

xs=np.zeros(k, dtype=float)+4.187374873069311

ys=np.zeros(k, dtype=float)+-9.109341689102738

vxs=np.zeros(k, dtype=float)+(4.767657694842040E-03*365)

vys=np.zeros(k, dtype=float)+(2.317175381750079E-03*365)

#initial conditions for uranus at midnight March 22 2020 taken from jpl website

xu=np.zeros(k, dtype=float)+1.603877632457955E+01

yu=np.zeros(k, dtype=float)+1.162370493399320E+01

vxu=np.zeros(k, dtype=float)+(-2.330046599777639E-03*365)

vyu=np.zeros(k, dtype=float)+(3.003164415990516E-03*365)

#initial conditions for neptune at midnight March 22 2020 taken from jpl website

xn=np.zeros(k, dtype=float)+2.929477586205114E+01

yn=np.zeros(k, dtype=float)+-6.116798683553212

vxn=np.zeros(k, dtype=float)+(6.287377836294756E-04*365)

vyn=np.zeros(k, dtype=float)+(3.094327154770977E-03*365)




re=np.zeros(k)
rj=np.zeros(k)
rej=np.zeros(k)
rmer=np.zeros(k)
rv=np.zeros(k)
rmar=np.zeros(k)
rs=np.zeros(k)
ru=np.zeros(k)
rn=np.zeros(k)

for n in range (0,k-1):
    #orbital radii of planets
    re[n]=np.sqrt((xe[n]**2+ye[n]**2))

    rj[n]=np.sqrt((xj[n]**2+yj[n]**2))
    
    rmer[n]=np.sqrt((xmer[n]**2+ymer[n]**2))
    
    rv[n]=np.sqrt((xv[n]**2+yv[n]**2))
    
    rmar[n]=np.sqrt((xmar[n]**2+ymar[n]**2))
    
    rs[n]=np.sqrt((xs[n]**2+ys[n]**2))
    
    ru[n]=np.sqrt((xu[n]**2+yu[n]**2))
    
    rn[n]=np.sqrt((xn[n]**2+yn[n]**2))
    
    #distance between each planet
    rej=np.sqrt( (xj[n]-xe[n])**2 + (yj[n]-ye[n])**2)
    
    rem=np.sqrt( (xe[n]-xmer[n])**2 + (ye[n]-ymer[n])**2)
    
    rjm=np.sqrt( (xj[n]-xmer[n])**2 + (yj[n] + xmer[n])**2)
    
    rev=np.sqrt( (xe[n]-xv[n])**2 + (ye[n]-yv[n])**2)
    
    rjv=np.sqrt( (xj[n]-xv[n])**2 + (yj[n]-yv[n])**2)
    
    rmv=np.sqrt( (xv[n]-xmer[n])**2 + (yv[n]-ymer[n])**2)
    
    remar=np.sqrt( (xe[n]-xmar[n])**2 + (ye[n]-ymar[n])**2)
    
    rjmar=np.sqrt( (xj[n]-xmar[n])**2 + (yj[n]-ymar[n])**2)
    
    rmermar=np.sqrt( (xmer[n]-xmar[n])**2 + (ymer[n]-ymar[n])**2)
    
    rvmar=np.sqrt( (xv[n]-xmar[n])**2 + (yv[n]-ymar[n])**2)
    
    res=np.sqrt( (xe[n]-xs[n])**2 + (ye[n]-ys[n])**2)
    
    rms=np.sqrt( (xmer[n]-xs[n])**2 + (ymer[n]-ys[n])**2)
    
    rjs=np.sqrt( (xj[n]-xs[n])**2 + (yj[n]-ys[n])**2)
    
    rvs=np.sqrt( (xv[n]-xs[n])**2 + (yv[n]-ys[n])**2)
    
    rsmar=np.sqrt( (xmar[n]-xs[n])**2 + (ymar[n]-ys[n])**2)
    
    reu=np.sqrt( (xu[n]-xe[n])**2 + (yu[n]-ye[n])**2)
    
    rum=np.sqrt( (xu[n]-xmer[n])**2 + (yu[n]-ymer[n])**2)
    
    ruv=np.sqrt( (xu[n]-xv[n])**2 + (yu[n]-yv[n])**2)
    
    rju=np.sqrt( (xu[n]-xj[n])**2 + (yu[n]-yj[n])**2)
    
    rsu=np.sqrt( (xu[n]-xs[n])**2 + (yu[n]-ys[n])**2)
    
    rumar=np.sqrt( (xmar[n]-xu[n])**2 + (ymar[n]-yu[n])**2)
    
    ren=np.sqrt( (xn[n]-xe[n])**2 + (yn[n]-ye[n])**2)
    
    rnm=np.sqrt( (xn[n]-xmer[n])**2 + (yn[n]-ymer[n])**2)
    
    rnv=np.sqrt( (xn[n]-xv[n])**2 + (yn[n]-yv[n])**2)
    
    rjn=np.sqrt( (xn[n]-xj[n])**2 + (yn[n]-yj[n])**2)
    
    rsn=np.sqrt( (xn[n]-xs[n])**2 + (yn[n]-ys[n])**2)
    
    rnmar=np.sqrt( (xn[n]-xmar[n])**2 + (yn[n]-ymar[n])**2)
    
    run=np.sqrt( (xn[n]-xu[n])**2 + (yn[n]-yu[n])**2)
    
    
    
    
    
    
    
    #velocity difference equation in x, y planes for each planet
    vxe[n+1]= vxe[n] - ((z*xe[n])/(re[n]**3))*dt - (dt*MJS*(xe[n]-xj[n]))/(rej**3) - (dt*MMS*(xe[n]-xmer[n])/rem**3) - (dt*MVS*(xe[n]-xv[n])/rev**3) - (dt*MMarS*(xe[n]-xmar[n])/remar**3) - (dt*MSS*(xe[n]-xs[n])/res**3) - (dt*MUS*(xe[n]-xu[n])/reu**3) - (dt*MNS*(xe[n]-xn[n])/ren**3)
    
    vye[n+1]= vye[n] - ((z*ye[n])/(re[n]**3))*dt - (dt*MJS*(ye[n]-yj[n]))/(rej**3) - (dt*MMS*(ye[n]-ymer[n])/rem**3) - (dt*MVS*(ye[n]-yv[n])/rev**3) - (dt*MMarS*(ye[n]-ymar[n])/remar**3) - (dt*MSS*(ye[n]-ys[n])/res**3) - (dt*MUS*(ye[n]-yu[n])/reu**3) - (dt*MNS*(ye[n]-yn[n])/ren**3)
    
    vxj[n+1]= vxj[n] - ((z*xj[n])/(rj[n]**3))*dt - (dt*MES*(xj[n]-xe[n]))/(rej**3) - (dt*MMS*(xj[n]-xmer[n])/rjm**3) - (dt*MVS*(xj[n]-xv[n])/rjv**3) - (dt*MMarS*(xj[n]-xmar[n])/rjmar**3) - (dt*MSS*(xj[n]-xs[n])/rjs**3) - (dt*MUS*(xj[n]-xu[n])/rju**3) - (dt*MNS*(xj[n]-xn[n])/rjn**3)
    
    vyj[n+1]= vyj[n] - ((z*yj[n])/(rj[n]**3))*dt - (dt*MES*(yj[n]-ye[n]))/(rej**3) - (dt*MMS*(yj[n]-ymer[n])/rjm**3) - (dt*MVS*(yj[n]-yv[n])/rjv**3) - (dt*MMarS*(yj[n]-ymar[n])/rjmar**3) - (dt*MSS*(yj[n]-ys[n])/rjs**3) - (dt*MUS*(yj[n]-yu[n])/rju**3) - (dt*MNS*(yj[n]-yn[n])/rjn**3)
    
    vxmer[n+1]= vxmer[n] - ((z*xmer[n])/(rmer[n]**3))*dt - (dt*MES*(xmer[n]-xe[n]))/(rem**3) - (dt*MJS*(xmer[n]-xj[n])/rjm**3) - (dt*MVS*(xmer[n]-xv[n])/rmv**3) - (dt*MMarS*(xmer[n]-xmar[n])/rmermar**3) - (dt*MSS*(xmer[n]-xs[n])/rms**3) - (dt*MUS*(xmer[n]-xu[n])/rum**3) - (dt*MNS*(xmer[n]-xn[n])/rnm**3)
    
    vymer[n+1]= vymer[n] - ((z*ymer[n])/(rmer[n]**3))*dt - (dt*MES*(ymer[n]-ye[n]))/(rem**3) - (dt*MJS*(xmer[n]-xj[n])/rjm**3) - (dt*MVS*(ymer[n]-yv[n])/rmv**3) - (dt*MMarS*(ymer[n]-ymar[n])/rmermar**3) - (dt*MSS*(ymer[n]-ys[n])/rms**3) - (dt*MUS*(ymer[n]-yu[n])/rum**3) - (dt*MNS*(ymer[n]-yn[n])/rnm**3)
    
    vxv[n+1]= vxv[n] - ((z*xv[n])/(rv[n]**3))*dt - (dt*MJS*(xv[n]-xj[n])/rjv**3) - (dt*MES*(xv[n]-xe[n]))/(rev**3) - (dt*MMS*(xv[n]-xmer[n])/rmv**3) - (dt*MMarS*(xv[n]-xmar[n])/rvmar**3) - (dt*MSS*(xv[n]-xs[n])/rvs**3) - (dt*MUS*(xv[n]-xu[n])/ruv**3) - (dt*MNS*(xv[n]-xn[n])/rnv**3)
    
    vyv[n+1]= vyv[n] - ((z*yv[n])/(rv[n]**3))*dt - (dt*MJS*(yv[n]-yj[n])/rjv**3) - (dt*MES*(yv[n]-ye[n]))/(rev**3) - (dt*MMS*(yv[n]-ymer[n])/rmv**3) - (dt*MMarS*(yv[n]-ymar[n])/rvmar**3) - (dt*MSS*(yv[n]-ys[n])/rvs**3) - (dt*MUS*(yv[n]-yu[n])/ruv**3) - (dt*MNS*(yv[n]-yn[n])/rnv**3)
    
    vxmar[n+1]= vxmar[n] - ((z*xmar[n])/(rmar[n]**3))*dt - (dt*MJS*(xmar[n]-xj[n])/rjmar**3) - (dt*MES*(xmar[n]-xe[n]))/(remar**3) - (dt*MMS*(xmar[n]-xmer[n])/rmermar**3) - (dt*MVS*(xmar[n]-xv[n])/rjmar**3) - (dt*MSS*(xmar[n]-xs[n])/rsmar**3) - (dt*MUS*(xmar[n]-xu[n])/rumar**3) - (dt*MNS*(xmar[n]-xn[n])/rnmar**3)
    
    vymar[n+1]= vymar[n] - ((z*ymar[n])/(rmar[n]**3))*dt - (dt*MJS*(ymar[n]-yj[n])/rjmar**3) - (dt*MES*(ymar[n]-ye[n]))/(remar**3) - (dt*MMS*(ymar[n]-ymer[n])/rmermar**3) - (dt*MVS*(ymar[n]-yv[n])/rjmar**3) - (dt*MSS*(ymar[n]-ys[n])/rsmar**3) - (dt*MUS*(ymar[n]-yu[n])/rumar**3) - (dt*MNS*(ymar[n]-yn[n])/rnmar**3)
    
    vxs[n+1]= vxs[n] - ((z*xs[n])/(rs[n]**3))*dt - (dt*MMarS*(xs[n]-xmar[n])/rsmar**3) - (dt*MJS*(xs[n]-xj[n])/rjs**3) - (dt*MMS*(xs[n]-xmer[n])/rms**3) - (dt*MES*(xs[n]-xe[n]))/(res**3) - (dt*MVS*(xs[n]-xv[n])/rvs**3) - (dt*MUS*(xs[n]-xu[n])/rsu**3) - (dt*MNS*(xs[n]-xn[n])/rsn**3)
    
    vys[n+1]= vys[n] - ((z*ys[n])/(rs[n]**3))*dt - (dt*MMarS*(ys[n]-ymar[n])/rsmar**3) - (dt*MJS*(ys[n]-yj[n])/rjs**3) - (dt*MMS*(ys[n]-ymer[n])/rms**3) - (dt*MES*(ys[n]-ye[n]))/(res**3) - (dt*MVS*(ys[n]-yv[n])/rvs**3) - (dt*MUS*(ys[n]-yu[n])/rsu**3) - (dt*MNS*(ys[n]-yn[n])/rsn**3)
    
    vxu[n+1]= vxu[n] - ((z*xu[n])/(ru[n]**3))*dt - (dt*MMarS*(xu[n]-xmar[n])/rumar**3) - (dt*MJS*(xu[n]-xj[n])/rju**3) - (dt*MMS*(xu[n]-xmer[n])/rum**3) - (dt*MES*(xu[n]-xe[n]))/(reu**3) - (dt*MVS*(xu[n]-xv[n])/ruv**3) - (dt*MSS*(xu[n]-xmar[n])/rumar**3) - (dt*MNS*(xu[n]-xn[n])/run**3)
    
    vyu[n+1]= vyu[n] - ((z*yu[n])/(ru[n]**3))*dt - (dt*MMarS*(yu[n]-ymar[n])/rumar**3) - (dt*MJS*(yu[n]-yj[n])/rju**3) - (dt*MMS*(yu[n]-ymer[n])/rum**3) - (dt*MES*(yu[n]-ye[n]))/(reu**3) - (dt*MVS*(yu[n]-yv[n])/ruv**3) - (dt*MSS*(yu[n]-ymar[n])/rumar**3) - (dt*MNS*(yu[n]-yn[n])/run**3)
    
    vxn[n+1]= vxn[n] - ((z*xu[n])/(rn[n]**3))*dt - (dt*MMarS*(xn[n]-xmar[n])/rnmar**3) - (dt*MJS*(xn[n]-xj[n])/rjn**3) - (dt*MMS*(xn[n]-xmer[n])/rnm**3) - (dt*MES*(xn[n]-xe[n]))/(ren**3) - (dt*MVS*(xn[n]-xv[n])/rnv**3) - (dt*MSS*(xn[n]-xmar[n])/rnmar**3) - (dt*MUS*(xn[n]-xu[n])/run**3)
    
    vyn[n+1]= vyn[n] - ((z*yu[n])/(rn[n]**3))*dt - (dt*MMarS*(yn[n]-ymar[n])/rnmar**3) - (dt*MJS*(yn[n]-yj[n])/rjn**3) - (dt*MMS*(yn[n]-ymer[n])/rnm**3) - (dt*MES*(yn[n]-ye[n]))/(ren**3) - (dt*MVS*(yn[n]-yv[n])/rnv**3) - (dt*MSS*(yn[n]-ymar[n])/rnmar**3) - (dt*MUS*(yn[n]-yu[n])/run**3)
    

    #position equations for each planet in x, y plane
    xe[n+1]= xe[n] + (vxe[n+1]*dt)

    ye[n+1]= ye[n] + (vye[n+1]*dt)

    xj[n+1]= xj[n] + (vxj[n+1]*dt)

    yj[n+1]= yj[n] + (vyj[n+1]*dt)
    
    xmer[n+1]=xmer[n] + (vxmer[n+1]*dt)
    
    ymer[n+1]=ymer[n] + (vymer[n+1]*dt)
    
    xv[n+1]= xv[n] + (vxv[n+1]*dt)
    
    yv[n+1]= yv[n] + (vyv[n+1]*dt)
    
    xmar[n+1]= xmar[n] + (vxmar[n+1]*dt)
    
    ymar[n+1]= ymar[n] + (vymar[n+1]*dt)
    
    xs[n+1]= xs[n] + (vxs[n+1]*dt)
    
    ys[n+1]= ys[n] + (vys[n+1]*dt)
    
    xu[n+1]= xu[n] + (vxu[n+1]*dt)
    
    yu[n+1]= yu[n] + (vyu[n+1]*dt)
    
    xn[n+1]= xn[n] + (vxn[n+1]*dt)
    
    yn[n+1]= yn[n] + (vyn[n+1]*dt)
    
    

plt.plot(xe, ye, xj, yj, xmer, ymer, xv, yv, xmar, ymar, xs, ys, xu, yu, xn, yn)
plt.xlabel('Position in x-domain')
plt.ylabel('Position in y-domain')
plt.title('Mercury, Earth and Jupiter orbiting the sun over 1 year')
```
